```python
import seaborn as sns
import statsmodels.formula.api as smf

df = sns.load_dataset('mpg').dropna()
print(df.head())
print(df.shape)
```

```python
model = smf.ols('mpg ~ weight', data=df).fit()
print(model.summary())
```

```python
model2 = smf.ols('mpg ~ weight + horsepower + cylinders', data=df).fit()
print(model2.summary())
```

```python
model3 = smf.ols('mpg ~ weight + model_year', data=df).fit()
print(model3.summary())
```
